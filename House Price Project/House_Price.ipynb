```python
categorical_features=[]
for i in data_set.columns:
    if data_set[i].dtype == 'object':
        categorical_features.append(i)
print(categorical_features)
```

```python
len(categorical_features)
```

```python
test_data_set=pd.read_csv("final_test.csv")
```

```python
test_data_set.head()
```

```python
final_data_set=pd.concat([data_set, test_data_set],axis=0)
```

```python
final_data_set.shape
```

```python
def onehot_mulcol(c_f):
    df_final=final_data_set
    j=0
    for i in c_f:
        print(i)
        df=pd.get_dummies(final_data_set[i],drop_first=True)
        final_data_set.drop([i],axis=1,inplace=True)
        if j==0:
            df_final=df.copy()
        else:
            df_final=pd.concat([df_final,df],axis=1)
        j+=1
    df_final=pd.concat([final_data_set,df_final],axis=1)
    return df_final
    
```

```python
final_data_set=onehot_mulcol(categorical_features)
```

```python
final_data_set.drop(['Id'],axis=1,inplace=True)
```

```python
final_data_set=final_data_set.loc[:,~final_data_set.columns.duplicated()]
```

```python
train=final_data_set.iloc[:1422,:]
test=final_data_set.iloc[1422:,:]
```

```python
test.drop(['SalePrice'], axis=1, inplace=True)
```

```python
X_train=train.drop(['SalePrice'],axis=1)
y_train=train['SalePrice']
```

```python
import xgboost
regressor=xgboost.XGBRegressor()
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
booster=['gbtree','gblinear']
base_score=[0.25,0.5,0.75,1]
n_estimators = [100, 500, 900, 1100, 1500]
max_depth = [2, 3, 5, 10, 15]
learning_rate=[0.05,0.1,0.15,0.20]
min_child_weight=[1,2,3,4]

hyperparameter_grid = {
    'n_estimators': n_estimators,
    'max_depth':max_depth,
    'learning_rate':learning_rate,
    'min_child_weight':min_child_weight,
    'booster':booster,
    'base_score':base_score
    }
```

```python
random_cv = RandomizedSearchCV(estimator=regressor,
            param_distributions=hyperparameter_grid,
            cv=5, n_iter=50,
            scoring = 'neg_mean_absolute_error',n_jobs = 4,
            verbose = 5, 
            return_train_score = True,
            random_state=42)
```

```python
random_cv.fit(X_train,y_train)
```

```python
random_cv.best_estimator_
```

```python
regressor=xgboost.XGBRegressor(base_score=0.25, booster='gbtree', colsample_bylevel=1,
             colsample_bynode=1, colsample_bytree=1, gamma=0, gpu_id=-1,
             importance_type='gain', interaction_constraints='',
             learning_rate=0.1, max_delta_step=0, max_depth=2,
             min_child_weight=1, missing=None, monotone_constraints='()',
             n_estimators=900, n_jobs=0, num_parallel_tree=1,
             objective='reg:squarederror', random_state=0, reg_alpha=0,
             reg_lambda=1, scale_pos_weight=1, subsample=1, tree_method='exact',
             validate_parameters=1, verbosity=None)
```

```python
regressor.fit(X_train,y_train)
```
